```python
#对于多组学数据进行数据对齐操作
X_cnv = pd.read_csv("X_cnv.csv",sep=",")
X_RNA = pd.read_csv("X_RNA.csv",sep=",")
number = 0
for i in range(0,len(X_cnv.columns)):
    for j in range(0,len(X_RNA.columns)):
        if(X_cnv.columns[i] == X_RNA.columns[j]):
            for h in range(0,len(Methylation_name)):
                if(X_cnv.columns[i] == Methylation_name[h]):
                    number = number+1
#print(number)    
select_cnv = np.zeros((len(X_cnv[X_cnv.columns[0]]),number))
select_RNA = np.zeros((len(X_RNA[X_RNA.columns[0]]),number))
select_Methylation = np.zeros((len(X_Methylation[X_Methylation.columns[0]]),number))
Y_select = np.zeros((number,1))
X_cnv_select = pd.DataFrame(select_cnv)
X_RNA_select = pd.DataFrame(select_RNA)
X_Methylation_select = pd.DataFrame(select_Methylation)
Y = pd.DataFrame(Y_select)
#print(X_RNA_select)
number = 0
for i in range(0,len(X_cnv.columns)):
    for j in range(0,len(X_RNA.columns)):
        if(X_cnv.columns[i] == X_RNA.columns[j]):
            for h in range(0,len(Methylation_name)):
                if(X_cnv.columns[i] == Methylation_name[h]):
                    X_cnv_select[X_cnv_select.columns[number]] = X_cnv[X_cnv.columns[i]]
                    X_RNA_select[X_RNA_select.columns[number]] = X_RNA[X_RNA.columns[j]]
                    X_Methylation_select[X_Methylation_select.columns[number]] = X_Methylation[X_Methylation.columns[h]]
                    Y.iloc[number] = Y_cnv.iloc[i]
                    number = number+1
print(X_cnv_select)
print(X_RNA_select)
print(X_Methylation_select)
print(Y)
```

```python
re_cnv = np.zeros((number,number))
relation_cnv = pd.DataFrame(re_cnv)
#relation_cnv.columns = X_cnv_select.columns
for i in range(0,number):
    for j in range(0,number):
        relation_cnv.iloc[i][j] = X_cnv_select[X_cnv_select.columns[i]].corr(X_cnv_select[X_cnv_select.columns[j]],method='kendall')
print(relation_cnv)
```

```python
re_RNA = np.zeros((len(X_RNA_select.columns),len(X_RNA_select.columns)))
relation_RNA = pd.DataFrame(re_RNA)
relation_RNA.columns = X_RNA_select.columns
for i in range(0,len(X_RNA_select.columns)):
    for j in range(0,len(X_RNA_select.columns)):
        relation_RNA.iloc[i][j] = X_RNA_select[X_RNA_select.columns[i]].corr(X_RNA_select[X_RNA_select.columns[j]],method='kendall')
print(relation_RNA)
```

```python
re_Methylation = np.zeros((len(X_Methylation_select.columns),len(X_Methylation_select.columns)))
relation_Methylation = pd.DataFrame(re_Methylation)
relation_Methylation.columns = X_Methylation_select.columns
for i in range(0,len(X_Methylation_select.columns)):
    for j in range(0,len(X_Methylation_select.columns)):
        relation_Methylation.iloc[i][j] = X_Methylation_select[X_Methylation_select.columns[i]].corr(X_Methylation_select[X_Methylation_select.columns[j]],method='kendall')
print(relation_Methylation)
```

```python
relation_cR = pd.concat([relation_cnv,relation_RNA],axis=1)
relation = pd.concat([relation_cR,relation_Methylation],axis=1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(relation, Y,test_size=0.2)
print (X_train)
print (y_train)
print ('... Done')
```

```python
import os
from sklearn import metrics
import tensorflow as tf
from tensorflow.compat.v1 import ConfigProto
from tensorflow.compat.v1 import InteractiveSession
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score   # 准确率
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
config = ConfigProto()
config.gpu_options.allow_growth = True
sess = InteractiveSession(config=config)
print(tf.__version__)

#X_train,X_test,y_train,y_test
X_train = X_train.as_matrix()
X_test = X_test.as_matrix()
Xtrain = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
#Xtrain = Xtrain / 255.0
Xtext = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
model = tf.keras.models.Sequential([
  tf.keras.layers.Conv1D(filters=128, kernel_size=3, input_shape=(X_train.shape[1], 1)),
  #tf.keras.layers.MaxPooling1D(pool_size=2),
  tf.keras.layers.MaxPooling1D(pool_size=5),
  #tf.keras.layers.Conv1D(filters=64, kernel_size=3),
  tf.keras.layers.Conv1D(filters=64, kernel_size=3),
  tf.keras.layers.MaxPooling1D(pool_size=5),
  tf.keras.layers.Conv1D(filters=16, kernel_size=3),
  tf.keras.layers.MaxPooling1D(pool_size=5),
  tf.keras.layers.BatchNormalization(),
  #tf.keras.layers.BatchNormalization(),
  #tf.keras.layers.BatchNormalization(),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(512, activation='relu'),
  #tf.keras.layers.Dropout(0.2),
  #tf.keras.layers.Dense(256, activation='relu'),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(2, activation='softmax')
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
model.fit(Xtrain, y_train, batch_size=90, epochs=150,validation_data=(Xtext, y_test))
test_loss = model.evaluate(Xtext, y_test)

y = model.predict(Xtext)
print (metrics.classification_report(y_test,y[:, 1].round()))
```
