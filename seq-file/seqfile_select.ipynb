```python
#对于RNA-SEQ数据的head文件进行标记用于重新生成每个药物的seq组学数据文件
import pandas as pd
data = pd.read_csv("seqsample-name.csv",sep=",")
record = pd.read_csv("seqsample-name.csv",sep=",")
print (data)
```

```python
#重新建立seq文件的head,每个数据集只需要执行一次
print("start")
for i in range(0,len(data["name1"])):
    for j in range(0,len(data["name2"])):
        if(data["name1"][i] == data["name2"][j]):
            record["name1"][i] = data["cell_line"][j]
record["name1"].to_csv("seq_name.csv")
print("end")
```

```python
#根据药物重新建立seq数据集，每个药物都需要执行，RNA-SEQV2为标记后的数据集
import pandas as pd
rna = pd.read_csv("RNA-SEQV2.csv",sep=",")
drug = pd.read_csv("../drug_1.csv",sep=",")
print(drug)
```

```python
#name为目标药物名称
list = [] ## 空列表
name = "FH535"
c = 0
for i in range(0,len(drug["DRUG_NAME"])):
    if(i%1000 == 0):
        print(i)
    if(drug["DRUG_NAME"][i] == name):
        c = c+1
    else:          
        list.append(i)
drug = drug.drop(list)
print(drug["DRUG_NAME"])
```

```python
#对于药物进行重新排序并将DATASET这一列赋值为Y
drug_1 = drug.sort_values(['LN_IC50'])
for i in range(0,c):
    drug_1["DATASET"][i] = i+1
drug_1.to_csv("drug_test.csv")
print(drug_1["DATASET"])
```

```python
drug_test = pd.read_csv("drug_test.csv",sep=",")
for i in range(0,c):
    drug_test["DATASET"][i] = i + 1
print(drug_test["DATASET"])
drug_test.to_csv("../drug_IC50.csv")
```

```python
#对于药物IC50信息进行二分类标记，drug_re为药物分类文件运算后结果
drug_bin = pd.read_csv("../drug_re.csv",sep=",",header = None)
drug_2 = pd.read_csv("drug_test.csv",sep=",")
for i in range(0,len(drug_2["CELL_LINE_NAME"])):
    drug_2["LN_IC50"][i] = drug_bin.iloc[i][1]
print(drug_2["LN_IC50"])
```

```python
#去掉RNAseq数据和药物数据中本次药物试验多余的样本
list = []
for h in range(0,len(drug_2["DATASET"])):
    drug_2["DATASET"][h] = 0
for i in range(0,len(rna.columns)):
    a = 0
    for j in range(0,len(drug_2["CELL_LINE_NAME"])):
        if(rna.columns[i] == drug_2["CELL_LINE_NAME"][j]):
            drug_2["DATASET"][j] = i +1
            a = 1
    if(a == 0):
        list.append(rna.columns[i])
rna = rna.drop(list,axis=1)
print(rna.columns) 
```

```python
list = []
drug_2 = drug_2.sort_values(['DATASET'])
#print(drug_2["DATASET"])
for i in range(0,len(drug_2["DATASET"])):
    if(drug_2["DATASET"][i] == 0):
        list.append(i)
drug_2 = drug_2.drop(list)
print(drug_2["DATASET"])
```

```python
#查看数据集此时的样本和分类标记
print(drug_2["CELL_LINE_NAME"])
print(drug_2["LN_IC50"])
drug_2.to_csv("../Sunitinib.csv")
rna.to_csv("../rna.csv")
#rna.to_csv("./Seliciclib/RNA_zen.csv",index = False)
#drug_2.to_csv("./Seliciclib/Vinblastine.csv",index = False)
```
